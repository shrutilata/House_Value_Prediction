```python
from sklearn.linear_model import LinearRegression
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
from sklearn import datasets ##imports dataset from scikit-Learn
boston = datasets.load_boston()
```

```python
#define the data/predictors as the pre-set feature names
df_x = pd.DataFrame(boston.data, columns=boston.feature_names)

#Put the target(housing value --MEDV) in another DataFrame
df_y = pd.DataFrame(boston.target)
```

```python
df_y
```

```python
df_x
```

```python
reg = LinearRegression()
```

```python
x_train,x_test,y_train,y_test = train_test_split(df_x,df_y,test_size=.2,random_state=4)
```

```python
reg.fit(x_train, y_train)
```

```python
x_test
```

```python
y_pred_val = reg.predict([[0.04560,0.0,13.89,1.0,0.5550,5.888,56.0,3.1121,5.0,276.0,16.4,392.80,13.5]])
```

```python
y_pred_val
```

```python
y_pred=reg.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
